# Cleaning the data

```python
#import libraies
import pandas as pd
import numpy as np
import os 
import matplotlib.pyplot as plt
import matplotlib as mpl
import seaborn as sns
import scipy
```

```python
path = r'/Users/ajahanderson/Documents/Book Data'
```

```python
df = pd.read_csv(os.path.join(path,'2 Data','Original Data','Books_selling_records.csv'),index_col = False)
```

```python
df.info()
```

```python
#remove columns
df = df.drop(columns = ['Quarter','Weekend','Morning','Afternoon','Night','Gender\nF=1665\nM=1815','Month','Year','Date.1',])
```

```python
#renaming columns
df.rename(columns = {'Total_\namount' : 'Total'}, inplace = True)
df.rename(columns = {'Profit Percentage\n (%)' : 'Profit %'}, inplace = True)
df.rename(columns = {'Day' : 'Day of week'}, inplace = True)
```

```python
#fixing grammar issues
df['Ship-State'] = df['Ship-State'].str.title()
df['Category'] = df['Category'].str.title()
df['Ship-State'] = df['Ship-State'].str.lower()
df['Ship-City'] = df['Ship-City'].str.lower()
df['Publication'] = df['Publication'].str.lower()

```

```python
#fixing grammar issues
df['Ship-State'] = df['Ship-State'].str.title()
df['Ship-City'] = df['Ship-City'].str.title()
df['Publication'] = df['Publication'].str.title()
df
```

```python
print(df.columns.tolist())
```

```python
df.rename(columns = {'Author ' : 'Author'}, inplace = True)
df.rename(columns = {'Online Sale Offers ' : 'Online Sale Offers'}, inplace = True)
```

```python
df['Author'] = df['Author'].str.lower()
```

```python
df['Author'] = df['Author'].str.title()
```

```python
df
```

```python
pd.set_option('display.max_columns', len(df))
```

```python
df
```

```python
#converting dtypes

df['Special Day'] = df['Special Day'].astype(str)
df['Online Sale Offers'] = df['Online Sale Offers'].astype(str)
df['Date'] = pd.to_datetime(df['Date'], format='%m/%d/%y')
df['Time'] = pd.to_datetime(df['Time']).dt.time

```

```python
#Changing to m and f so its easier to read
df['Gender'] = df['Gender'].replace({0: 'M', 1: 'F'})
```

```python
df.info()
```

```python
#checking for mixed data types
for col in df.columns.tolist():
  weird = (df[[col]].applymap(type) != df[[col]].iloc[0].apply(type)).any(axis = 1)
  if len (df[weird]) > 0:
    print (col)
```

```python
#Checking for missing values
df.isnull().sum()
```

```python
#Checking for missing values
df_dups = df[df.duplicated()]
```

```python
df_dups
```

```python
#removing columns
df.drop(columns=['Evening'], inplace=True)
df.drop(columns=['Purchase-date'], inplace=True)
```
