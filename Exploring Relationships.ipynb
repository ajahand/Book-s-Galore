## Correlations:

```python
pd.set_option('display.float_format',lambda x:'%.1f'%x)
```

```python
sub.corr()
```

```python
f,ax = plt.subplots(figsize=(10,10))

# Create the correlation heatmap in seaborn by applying a heatmap onto the correlation matrix and the subplots defined above.
corr = sns.heatmap(sub.corr(), annot = True, ax = ax) # The `annot` argument allows the plot to 
#place the correlation coefficients onto the heatmap.
```

At first glance, it's evident that the light-colored squares indicate a positive correlation. Many currency-related variables, such as cost prices and profit, are positively linked to other currency values.

### Categorical Plots:
